```python
url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/main/TelecomX_Data.json"
df = pd.read_json(url)
df.head()
```

```python
print(df.columns)
```

```python
# Normalizar colunas aninhadas
df_expanded = pd.concat([
    df.drop(['customer', 'phone', 'internet', 'account'], axis=1),
    pd.json_normalize(df['customer']),
    pd.json_normalize(df['phone']),
    pd.json_normalize(df['internet']),
    pd.json_normalize(df['account'])
], axis=1)

# Verificar se 'TotalCharges' agora existe
print(df_expanded.columns)
```

```python
df_expanded['Charges.Total'] = pd.to_numeric(df_expanded['Charges.Total'], errors='coerce')
df_expanded['Charges.Total'] = df_expanded['Charges.Total'].fillna(0)
```

```python
# Criar coluna com valor diário da conta
df_expanded['Contas_Diarias'] = df_expanded['Charges.Monthly'] / 30

# Verificar resultado
print(df_expanded[['Charges.Monthly', 'Contas_Diarias']].head())
```

```python
df_expanded.groupby('Churn')['Contas_Diarias'].mean()
```

```python
df_expanded['Partner'] = df_expanded['Partner'].map({'Yes':1, 'No':0})
df_expanded['Dependents'] = df_expanded['Dependents'].map({'Yes':1, 'No':0})
df_expanded['PaperlessBilling'] = df_expanded['PaperlessBilling'].map({'Yes':1, 'No':0})
```

```python
df.rename(columns={
    'MonthlyCharges':'Faturamento_Mensal',
    'TotalCharges':'Faturamento_Total',
    'Churn':'Evasao'
}, inplace=True)
```

```python
df_expanded['Contract'] = df_expanded['Contract'].replace({
    'Month-to-month':'Mensal',
    'One year':'Anual',
    'Two year':'Bienal'
})
```

```python
# Estatísticas descritivas básicas
df_expanded.describe()

# Média, mediana e desvio padrão de colunas relevantes
print("Média - Charges.Monthly:", df_expanded['Charges.Monthly'].mean())
print("Mediana - Charges.Monthly:", df_expanded['Charges.Monthly'].median())
print("Desvio padrão - Charges.Monthly:", df_expanded['Charges.Monthly'].std())

print("Média - Charges.Total:", df_expanded['Charges.Total'].mean())
print("Mediana - Charges.Total:", df_expanded['Charges.Total'].median())
print("Desvio padrão - Charges.Total:", df_expanded['Charges.Total'].std())

print("Média - Contas_Diarias:", df_expanded['Contas_Diarias'].mean())
print("Mediana - Contas_Diarias:", df_expanded['Contas_Diarias'].median())
print("Desvio padrão - Contas_Diarias:", df_expanded['Contas_Diarias'].std())
```

## RELATÓRIO FINAL

### **1. INTRODUÇÃO**
O objetivo desta análise é compreender os fatores que influenciam a evasão de clientes **(Churn)** na empresa TELECOM X, uma organização de telecomunicações.
A evasão representa um desafio estratégico, pois impacta diretamente na receita e na sustentabilidade do negócio.
Através da exploração dos dados, buscamos identificar padrões e variáveis que possam explicar por que alguns clientes cancelam seus contratos, enquanto outros permanecem.


###**2. Limpeza e Tratamentos dos Dados**

- Importação dos dados a partir do arquivo **TelecomX_Data.json.**
- Normalização de colunas aninhadas **(customer, phone, internet, account).**
- Conversão de variáveis numéricas que estavam como texto (**TotalCharges** → float).
- Tratamento de valores nulos em **TotalCharges** (substituídos por 0).
- Remoção de espaços em branco em colunas categóricas.
- Padronização de categorias (**No internet service → No**).
- Conversão da variável alvo **Churn** para binária **(Yes=1, No=0).**
- Criação da coluna **Contas_Diarias** (valor diário baseado no faturamento mensal).

###**3. Análise Exploratória de Dados**
- **Distribuição da variável Churn:** gráficos de barras e pizza mostraram que a maioria dos clientes permanece, mas há uma proporção significativa de evasão.

 **Variáveis categóricas:**
- Contratos mensais apresentam maior taxa de evasão.
- O método de pagamento influencia: clientes com cobrança eletrônica têm maior tendência a cancelar.
- Gênero não mostrou diferença significativa.

 **Variáveis numéricas:**
 - Clientes com menor tempo de contrato **(tenure)** e menor gasto acumulado **(TotalCharges)** tendem a evadir mais.
 - O faturamento mensal **(MonthlyCharges)** também mostrou associação com evasão, especialmente em valores mais altos.

###**4. Conclusões e Insights**

- Tempo de contrato é um dos fatores mais relevantes: clientes novos têm maior risco de evasão.

- Tipo de contrato influencia fortemente: contratos mensais são mais vulneráveis.

- Valor gasto: clientes com baixo gasto acumulado tendem a evadir, indicando que a fidelização ainda não ocorreu.

- Método de pagamento: pode ser um indicador de perfil de cliente mais propenso ao cancelamento.

###**5. Recomendações**

- Incentivar contratos de longo prazo: oferecer descontos ou benefícios para clientes que migram de contratos mensais para anuais/bienais.

- rogramas de fidelização: criar ações específicas para clientes novos (primeiros meses de contrato), reduzindo o risco de evasão inicial.

- Monitorar clientes com alto custo mensal: oferecer pacotes personalizados ou suporte adicional para evitar insatisfação.

- Revisar métodos de pagamento: analisar se clientes com determinados métodos (ex: cobrança eletrônica) precisam de comunicação diferenciada ou suporte extra.
